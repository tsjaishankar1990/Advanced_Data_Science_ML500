# Normal Distribution

The normal distribution is perhaps the most obiquitous and familiar distribution we encounter. It is also referred to as the Gaussian distribution, eponymously named after its discoverer, Gauss.

For univariate data (i.e. scalar, or one-dimensional data), it takes the familiar shape of a "bell curve" -- and so is referred by that common name.

In this lab, we will see how the variance parameter $\mathbf{\sigma^2}$ affects the shape of the curve. Recall that the $\mathbf{\mu}$ parameter specifies the mean of the distribution -- i.e. where the "peak" is centered or located.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics


def plot_normal(σ):
    # Plot between -10 and 10 with .001 steps.
    x_axis = np.arange(-50, 50, 0.01)
    μ = statistics.mean(x_axis)
    plt.plot(x_axis, norm.pdf(x_axis, μ, σ))


plt.title("Normal Distribution", size=20)
plot_normal(σ=10)
plot_normal(σ=5)
plot_normal(σ=2)
plot_normal(σ=1)

plt.legend([r"$\sigma^2:100$", r"$\sigma^2:25$", r"$\sigma^2:4$", r"$\sigma^2:1$"], loc="upper right")
plt.show()
```
