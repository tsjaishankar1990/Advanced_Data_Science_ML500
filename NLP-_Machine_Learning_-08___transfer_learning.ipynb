## Transfer learning

HuggingFace has a large repository of open-source models that have been pretrained for specific tasks. Quite often, the problem we would like to solve simply requires us to use one of these pretrained models.

However, if we need to solve a task for which a model does not exist in the huggingface repository, we can start with a pre-trained model trained on something that is somewhat close to our task (i.e. an adjacent task), and then apply transfer learning.

In other words, we can take labeled data for our specific task, and then train the pre-trained model and train it further (i.e. run a few epochs of training through our specific training dataset). 

Since we started with a pretrained model, the weights/parameters of the model have already been trained to a problem adjacent to ours. The parameter values form the **checkpoint** of the model. So more precisely, we are downloading and starting with a pre-trained model checkpoint.

Now, we do not need enormous amounts of data to further train the model to our task, since it is already trained to an adjacent task. In general, we expect the weights to go through small perturbations as we run a few more epochs.

This process of further training the model from a pre-trained checkpoint to one which is specific to our task is called **fine tuning**.

Therefore, transfer learning is the process of two steps:

* start with a **pretrained model checkpoint**, trained on an adjacent task

* **fine tune** the model with a few epochs of further training, using our task specific training dataset
